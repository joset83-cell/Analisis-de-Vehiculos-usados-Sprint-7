```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# Cargar el dataset
df = pd.read_csv('../vehicles_us.csv')
df.head()
```

```python
# Histograma de la columna odometer
fig_histogram = go.Figure(data=[go.Histogram(x=df['odometer'])])
fig_histogram.update_layout(
    title='Distribución del Odómetro',
    xaxis_title='Odómetro (millas)',
    yaxis_title='Frecuencia'
)
fig_histogram.show()
```

```python
# Gráfico de dispersión: Precio vs Año del modelo
fig_scatter = go.Figure(data=[go.Scatter(
    x=df['model_year'],
    y=df['price'],
    mode='markers',
    marker=dict(
        size=5,
        color=df['odometer'],
        colorscale='Viridis',
        showscale=True,
        colorbar=dict(title='Odómetro')
    )
)])
fig_scatter.update_layout(
    title='Precio vs Año del Modelo',
    xaxis_title='Año del Modelo',
    yaxis_title='Precio (USD)'
)
fig_scatter.show()
```
